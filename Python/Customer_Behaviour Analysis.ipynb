```python
# Replacing the null values with the median of the column grouped by its category
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
# Renaming columns according to snake casing for better readability and documentation
df.columns 
```

```python
# 1. Converting the column names to lower case
df.columns = df.columns.str.lower()
df.columns
```

```python
# 2. Replacing the space in between the column names with underscore
df.columns = df.columns.str.replace(' ','_')
df.columns
```

```python
# 3. Renaming the column
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
df.columns
```

```python
# Creating a new categorical column as "age_group" from the numerical column "age"
age_category = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels = age_category)
df[['age','age_group']].head(10)
```

```python
# Creating new numerical column as "purchase_frequency_days" from the categorical column "frequency_of_purchases"
frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
df[['purchase_frequency_days','frequency_of_purchases']].head(10)
```

```python
# Checking duplicate columns
df[['discount_applied','promo_code_used']].head(10)
```

```python
(df['discount_applied'] == df['promo_code_used']).all()
```

```python
# Dropping promo_code_used column since discount_applied column has same values
df = df.drop('promo_code_used', axis=1)
```
